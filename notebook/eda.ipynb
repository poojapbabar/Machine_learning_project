```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import missingno as msno
```

```python
data = pd.read_csv('adult.csv')
```

```python
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
msno.matrix(data)
```

```python
data.replace('?', np.nan, inplace = True) # Here we replace '?' with nan
data.head(20)
```

```python
data.dropna(inplace = True)
```

```python
ax = sns.countplot(data=data, x='income')
ax.set_xticklabels(['<=50K', '>50K']);
ax.set_xlabel('income');
ax.set_ylabel('Count');
ax.bar_label(ax.containers[0]);
```

```python
sns.histplot(data=data,x = 'age',bins = 25,kde = True,color = 'r')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data,x = 'education')
plt.title('Distribution of Education level')
plt.xlabel('education level')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(data=data,x = 'education', y = 'hours-per-week')
plt.title('Education vs Hours per week')
plt.xlabel('Education level')
plt.ylabel('Hours per week')
plt.xticks(rotation =45)
plt.show()
```

```python
sns.countplot(data=data,x = 'gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='income', hue='gender', multiple='stack')
plt.title('Salary Distribution by Gender')
plt.xlabel('income')
plt.ylabel('Count')
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import altair as alt
```

```python
px.bar(data['education'].value_counts(), x=data['education'].unique(), y=data['education'].value_counts(), labels={'x':'Education Level', 'y':'Count'}, title='Distribution of Education Levels')
```

```python
fig = px.scatter(data, x='capital-gain', y='capital-loss', color='income', title='Capital Gain vs. Capital Loss')
fig.show()
```

```python
plt.figure(figsize=(8, 8))
data['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
fig = px.bar(data, x='occupation', title='Occupation Distribution')
fig.show()
```

```python
# Plot 16: Matplotlib - Line Plot
plt.figure(figsize=(10, 6))
data.groupby('education')['hours-per-week'].mean().plot(kind='line', marker='o')
plt.title('Average Hours per Week by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Hours per Week')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 15: Plotly - Box Plot
fig = px.box(data, x='race', y='age', color='income', title='Age Distribution by Race')
fig.show()
```
